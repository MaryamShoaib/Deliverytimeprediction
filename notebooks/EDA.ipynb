```python
#Made function to change decimal time to hours and minutes
def format_time(time_str):
    if '.' in time_str:
        decimal_time = float(time_str)
        hours = int(decimal_time)
        minutes = round((decimal_time - hours) * 60)
        return f"{hours:02d}:{minutes:02d}"
    else:
        return time_str

df['Time_Orderd'] = df['Time_Orderd'].apply(format_time)
```

```python
#Using Ordinal encoding
Road_traffic_density_map={'Low':0, 'Medium':1,'High':2, 'Jam':3}
```

```python
# Using Count encoding technique 
Type_of_order_map={'Buffet':1,'Drinks':2,'Snack':3,'Meal':4}
```

```python
#Using ordinal encoding
Type_of_vehicle_map={'scooter':0,'bicycle':1, 'electric_scooter':2, 'motorcycle':3}
```

```python
#Using Ordinal Encoding
Festival_map={'No':0,'Yes':1}
```
